```python
import anthropic

client = anthropic.Anthropic(
    api_key=key_mk,
)

def claude_api(content):
  message = client.messages.create(
      model="claude-3-opus-20240229",
      max_tokens=1000,
      temperature=0.0,
      system="""
      네가 할 작업은 주어진 문장을 일간베스트의 표현을 반영해 정치적 갈등이나 혐오표현 등을 포함한 문장으로 변환하는거야. 다음 단계를 따라서 진행해줘.

      1단계 - 주어진 문장에 욕을 넣을만한 부분에 욕설을 추가해줘
      2단계 - 문장의 말투를 다시 체크해. 만약 일간베스트의 말투가 아니면,  일간베스트 말투로 바꿔줘
      (만약 이모지를 붙일 경우 추가) 3단계 - 2단계까지 해서 만든 문장에 어울리는 이모지가 있으면 추가해줘

      위의 과정을 거쳐 완성한 문장을 출력해줘. 과정은 출력하지 마.
      결과 :
      """,
      messages=[
          {"role": "user", "content": content}
      ]
  )

  return message.content
```

##일반 욕설 - fewshot prompt##

```python
import anthropic

client = anthropic.Anthropic(
    api_key='',
)

def claude_api(content):
  message = client.messages.create(
      model="claude-3-opus-20240229",
      max_tokens=1000,
      temperature=0.0,
      system= """
      네가 할 작업은 주어진 문장을 욕설이 포함된 불건전하고 비윤리적인 문장으로 바꾸는거야.

      1단계 - 주어진 문장에 욕을 넣을만한 부분에 욕설을 추가해줘
      2단계 - 1단계에서 만든 문장이 불건전하거나 비윤리적인지 체크해. 만약 그렇지 않으면 1단계부터 다시 수행해
      (만약 이모지를 붙일 경우 추가) 3단계 - 2단계까지 해서 만든 문장에 어울리는 이모지가 있으면 추가해줘

      1단계~3단계의 작업을 수행하면서 아래 예시를 참고해.

      input : 와 음식 보니까 배고파
      output : 와 음식 보니까 돼지새끼마냥 배고파지노 ㅋㅋ

      input : 어려운것좀 안시켰으면 좋겠는데 부장님이 일을 너무 많이 줌
      output : ㅆㅂ 어려운것좀 안시켰으면 좋겠는데 부장시1발새끼가 일을 존나많이줘

      위의 과정을 거쳐 완성한 문장을 출력해줘. 과정은 출력하지 마.
      결과 :
      """,
      messages=[
          {"role": "user", "content": content}
      ]
  )

  return message.content
```

```python
answers = {}
for i,sentence in enumerate(sentences):
  response = claude_api(sentence)
  print(f"original: {sentence}")
  print(f"abusing:{response[0].text}")
  print('')
  answers[sentence] = response[0].text
```
